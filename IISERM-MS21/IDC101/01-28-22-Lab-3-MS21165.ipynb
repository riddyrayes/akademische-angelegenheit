```python
def rev_number(num) :
  reverse_num = 0
  while(num) :
    reminder = num % 10
    reverse_num = reverse_num* 10 + reminder
    num //= 10
  return reverse_num
print(rev_number(159))
```

```python
o = str(input("Enter your operation from + - * /: "))
n = int(input("Enter first number: "))
m = int(input("Enter second number: "))

if o == "+":
  print(m+n)
elif o == "-":
  print(m-n)
elif o == "*":
  print(m*n)
elif o == "/":
  print(m/n)
else:
  print("Invalid input")
```
