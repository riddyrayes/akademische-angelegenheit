# Question 2

```python
def encrypt(s):
  o = ""
  for i in s:
      if ord(i) < ord("Y"):
        o += chr(ord(i)+2)
      else:
        o += chr(ord(i)+2-26)
  return o

def decrypt(s):
  o = ""
  for i in s:
      if ord(i) > ord("B"):
        o += chr(ord(i)-2)
      else:
        o += chr(ord(i)-2+26)
  return o


print(encrypt(str(input())))
print(decrypt(str(input())))
```

```python
import random

def pi_estimate(N):
  inside_circle=0

  for i in range(0,N):
    x_i=random.random()*2+-1
    y_i=random.random()*2+-1
    if (x_i**2+y_i**2)<=1:
      inside_circle+=1

  pi_estimate= (inside_circle/N)*4
  outside_circle=N-inside_circle

  return pi_estimate

print(pi_estimate(10000))
```
