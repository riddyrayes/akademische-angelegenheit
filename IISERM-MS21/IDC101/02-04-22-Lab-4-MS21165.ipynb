```py
## brute force algorithm

define prime_factors(n):
  let i be 2

    if i does not divide n:
      i = i+1
    else:
      n = quotient of n/i

```

````python
def pf(n):
  i = 2
  factors = []
  d={}
  while i ** 2 <= n:
      if n % i:
          i += 1
      else:
          n //= i
          factors.append(i)
  if n > 1:
      factors.append(n)
  for i in factors:
    d[i] = 0
  for i in factors:
    d[i] = d[i] + 1
  return d
  
n=1024*555
d=pf(n)
print(n,"=",end=" ")
for i in d:
  print("+",i,"^",d[i],end=" ")
````
