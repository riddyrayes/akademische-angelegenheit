```python
def bubbleSort(l):
  for i in range(len(l)):
    for j in range(0, len(l) - i - 1):
      if l[j] > l[j + 1]:
        l[j], l[j+1] = l[j+1], l[j]


b = [-2, 45, 0, 11, -9]
bubbleSort(b)
print(b)
```

```python
l = ["Adrian 55 65 90 88", "Jasuz 75 64 95 96", "Jeffrey 81 79 58 82", "Michal 79 59 80 65", "Piotr 85 91 78 80"]
s = "Math Biology Chemistry Physics"

## saving as list

for i in range(len(l)):
  l[i] = list(l[i].split(" "))
  for j in range(1, len(l[i])):
    l[i][j] = int(l[i][j])
print(l)

s = list(s.split(" "))

## calculating subject wise performance
t=[]

for j in range(1,len(l[i])):
  t.append(0)

for i in range(len(l)):
  for j in range(1,len(l[i])):
    if len(t)>j-1:
      t[j-1] += l[i][j]
    else:
      t.append(l[i][j])
for i in range(len(s)):
  print("Performance for",s[i],"is",t[i]/len(l))
```

```python
l = ["Adrian 55 65 90 88", "Jasuz 75 64 95 96", "Jeffrey 81 79 58 82", "Michal 79 59 80 65", "Piotr 85 91 78 80"]
s = "Math Biology Chemistry Physics"

## saving as dict
d={}
for i in range(len(l)):
  l[i] = list(l[i].split(" "))
  d[l[i][0]] = list(map(int, l[i][1:]))
print(d)
s = list(s.split(" "))

## calculating student wize average score
for i in d:
  print("Average score of",i,"is",sum(d[i])/len(d[i]))

## calculating subject wise performance
t=[]
for j in range(len(s)):
  t.append(0)
for i in d:
  for j in range(len(d[i])):
    t[j] += d[i][j]
for i in range(len(s)):
  print("Performance for",s[i],"is",t[i]/len(l))
```
